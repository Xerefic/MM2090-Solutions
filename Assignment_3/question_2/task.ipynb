```python
import numpy as np
np.random.seed(0)

from scipy import optimize

import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
def sigmoid(x_data, scale=5, explode=10):
    mean = np.mean(x_data*explode)
    y = (x_data*explode - mean)/scale
    return 1/(1+np.exp(-y))
```

```python
x_data = np.linspace(0, 5, num=100)
y_data = (sigmoid(x_data) + 0.05*np.random.normal(size=x_data.shape))
```

```python
plt.scatter(x_data, y_data, color='g')
plt.title("Fraction of Evolution as function of Time")
plt.xlabel("Time")
plt.ylabel("Fractional Growth")
plt.show()
```

```python
def fit_function(x, a, b, c):
    return 1/(1+np.exp((x*a-b)/c))
```

```python
params, params_covariance = optimize.curve_fit(fit_function, x_data, y_data, p0=[10,2,1])
```

```python
plt.scatter(x_data, y_data, c='g')
plt.plot(x_data, fit_function(x_data, params[0], params[1], params[2]), label='Fitted with Sigmoid')
plt.xlabel("Time")
plt.ylabel("Fractional Growth")
plt.legend(loc='best')
plt.show()
```
