```octave
img = imread("Original.jpg");
imshow(img)

alpha = 1.5;
beta = 0.2*255;
gamma = 1.2;
```

```octave
bright = min(img+beta,255);
contrast = min(img*alpha,255);
correct = min((img/255).^gamma * 255, 255);

mod = 255-min( ((alpha*img+beta)/255).^gamma * 255, 255);
imshow(mod)
```

```octave
detect = [-1,-1,-1;-1,8,-1;-1,-1,-1];
```

```octave
new = mod;
new(:,:,1) = conv2(img(:,:,1), detect, "same");
new(:,:,2) = conv2(img(:,:,2), detect, "same");
new(:,:,3) = conv2(img(:,:,3), detect, "same");
imshow(new)
```
